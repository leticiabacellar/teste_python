```python
# # Passo 8:  Calcular a média, mediana e população total das cidades da tabela 

df.groupby('Unidade federativa')

df.groupby('Unidade federativa')['População'].count()
df.groupby('Unidade federativa')['População'].max()
df.groupby('Unidade federativa')['População'].sum()

# df.groupby('Unidade federativa'): Cria grupos no DataFrame df com base nos valores únicos na coluna 'Unidade federativa'. 
# Essa operação retorna um objeto GroupBy que contém os grupos.

# df.groupby('Unidade federativa')['População'].count(): Conta o número de ocorrências da coluna 'População' dentro de cada
# grupo definido pela coluna 'Unidade federativa'. Retorna uma série que exibe o número de linhas (ocorrências) para cada grupo.

# df.groupby('Unidade federativa')['População'].max(): Encontra o valor máximo da coluna 'População' dentro de cada grupo 
# definido pela coluna 'Unidade federativa'. Retorna uma série que exibe o valor máximo para cada grupo.

# df.groupby('Unidade federativa')['População'].sum(): Soma os valores da coluna 'População' dentro de cada grupo definido
# pela coluna 'Unidade federativa'. Retorna uma série que exibe a soma dos valores para cada grupo.
```

```python
df_agg = df.groupby('Unidade federativa')['População'].agg(['count', 'max', 'sum'])
df_agg.columns = ['Contador', 'Pop máxima', 'Soma']
df_agg


# df.groupby('Unidade federativa')['População'].agg(['count', 'max', 'sum']): Agrupa os dados pelo valor da coluna 'Unidade 
# federativa' e aplica as operações agregadas especificadas dentro de agg(). No caso, as operações agregadas são 'count' 
# (contagem), 'max' (valor máximo) e 'sum' (soma) aplicadas à coluna 'População'. Isso resulta em um novo DataFrame chamado
# df_agg que contém os resultados dessas operações agregadas para cada grupo.

# df_agg.columns = ['Contador', 'Pop máxima', 'Soma']: Renomeia as colunas do DataFrame df_agg para 'Contador
# ', 'Pop máxima' e 'Soma', respectivamente. Essa linha de código atribui novos nomes às colunas resultantes das operações 
# agregadas.
```
